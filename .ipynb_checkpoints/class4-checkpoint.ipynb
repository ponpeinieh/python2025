```python
def calc_square_and_cube(number):
    square = number**2
    cube = number**3
    return [square, cube]

# 呼叫
num = 10
result = calc_square_and_cube(num)
print(f'{num}的平方:{result[0]}, 立方:{result[1]}')
```

```python
def calc_square_and_cube(number):
    square = number**2
    cube = number**3
    return {'square':square, 'cube':cube}

num = 10
result = calc_square_and_cube(num)
print(f'{num}的平方:{result['square']}, 立方:{result['cube']}')
```

```python
def x(a):
    print("in 函數 x中:",a)
    a += 1
    print("in 函數 x中:",a)
    return

def main():
    i = 100
    # 呼叫 x()
    x(i)
    print('in 函數 main中', i)

main()
```

```python
# 清單作爲參數型態
def add_block(blocks, new_block):
    blocks.append(new_block)
    return blocks # 可以不用回傳
# 呼叫
current_blocks = ['石頭', '磚塊','鑽石']
add_block(current_blocks, '黃金')
print(current_blocks)
```

```python
# 清單作爲參數型態
def add_block(blocks, new_block):
    blocks.append(new_block)
    return blocks # 可以不用回傳
# 呼叫
current_blocks = ['石頭', '磚塊','鑽石']
result = add_block(current_blocks[:], '黃金')
print(current_blocks)
print(result)
```
