```python
def calc_square_and_cube(number):
    square = number**2
    cube = number**3
    return [square, cube]

# 呼叫
num = 10
result = calc_square_and_cube(num)
print(f'{num}的平方:{result[0]}, 立方:{result[1]}')
```

```python
def calc_square_and_cube(number):
    square = number**2
    cube = number**3
    return {'square':square, 'cube':cube}

num = 10
result = calc_square_and_cube(num)
print(f'{num}的平方:{result['square']}, 立方:{result['cube']}')
```
