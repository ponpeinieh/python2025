```python
# 清單進階操作
shopping_list = []
# 新增元素: append() 往後面加入
shopping_list.append('麵包')
shopping_list.append('可樂')
shopping_list.append('橘子')
print(shopping_list)
item = input('請輸入下一個商品:')
shopping_list.append(item)
print(shopping_list)
# insert(在某個位置插入新的元素)
shopping_list.insert(2, '湯圓')
print(shopping_list)
```

```python
#刪除 (1). del (2). pop() (3). remove()
del shopping_list[1]
print(shopping_list)
```

```python
removed_item = shopping_list.pop(1)
print(f'被刪除的物品為{removed_item}')
print(f'剩下的物品為:{shopping_list}')
```

```python
# remove()是指定要刪除的内容
shopping_list.remove('麵包')
print(shopping_list)
```

```python
scores = {'John':90, 'Steven':85, 'Victoria':100}
# 尋訪
for name in scores.keys(): # 等同於 for name in scores:
    print(f'{name}的分數為{scores[name]}')
```

```python
# 直接取得每一個分數
for s in scores.values():
    print(s)
```

```python
# 一次取得key與value
for name,s in scores.items():
    print(f'{name}的分數為{s}!')
```

```python
# for loop (迴圈)
# 重複n次
for i in range(100): # 從0開始, 到100-1(99) -->100個循環
    print(i, end=',')
```

```python
# range(start, end, step) 
# start: 開始(預設為0)
# end: 結束(不包含)
# step: 間隔(預設為1)
for i in range(1,100,2):
    print(i, end=',')
```
