```python
def divide(a,b):
    try:
        return a/b
    except:
        print('發生不預期的錯誤!')
    return None

a = int(input('請輸入第一個數字:'))
b = int(input('請輸入第二個數字:'))
print(divide(a,b))
print('完成')
```

```python
def divide(a,b):
    try:
        return a/b
    except ZeroDivisionError:
        # print('發生不預期的錯誤!')
        print('除數不能為0!')
    return None
while True:
    try:
        a = int(input('請輸入第一個數字:'))
        b = int(input('請輸入第二個數字:'))
        break
    except ValueError:
        print('輸入的數值錯誤!請重新輸入!')
print(divide(a,b))
print('完成')
```

```python
def divide(a,b):
    return a/b
    
while True:
    try:
        a = int(input('請輸入第一個數字:'))
        b = int(input('請輸入第二個數字:'))
        print(divide(a,b))
        break
    except ValueError:
        print('輸入的數值錯誤!請重新輸入!')
    except ZeroDivisionError:
        print('除以0的錯誤!')

print('完成')
```

```python
# TypeError (型態錯誤)
def divide(a,b):
    return a/b
divide('a',10)
```

```python
def access_list(lst, index):
    if index<len(lst):
        return lst[index]
    else:
        return None

my_list = [1,2,3,4,5]
print(access_list(my_list,5))
```

```python
# KeyError 鍵值錯誤
def access_dict(dct, key):
    if key in dct:
        return dct[key]
    else:
        return None

my_dict = {'apple':10, 'orange':5, 'grapes':12}
print(access_dict(my_dict, 'apple'))
print(access_dict(my_dict, 'banana'))
```

```python
# KeyError 鍵值錯誤
def access_dict(dct, key):
    return dct.get(key,None)

my_dict = {'apple':10, 'orange':5, 'grapes':12}
print(access_dict(my_dict, 'apple'))
print(access_dict(my_dict, 'banana'))
```

```python
# 自訂錯誤(例外)型態
class NegativeValueError(Exception):
    pass # 沒有内容
# 丟出(raise 產生)例外的物件

def calc_bmi(height, weight):
    if height<0 or weight<0:
        raise NegativeValueError('身高及體重不能為負數!')
    return weight/height**2

h = float(input('身高:'))
w = float(input('體重:'))
print(calc_bmi(h,w))
```
