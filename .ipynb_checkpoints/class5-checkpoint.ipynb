```python
def divide(a,b):
    try:
        return a/b
    except:
        print('發生不預期的錯誤!')
    return None

a = int(input('請輸入第一個數字:'))
b = int(input('請輸入第二個數字:'))
print(divide(a,b))
print('完成')
```

```python
def divide(a,b):
    try:
        return a/b
    except ZeroDivisionError:
        # print('發生不預期的錯誤!')
        print('除數不能為0!')
    return None
while True:
    try:
        a = int(input('請輸入第一個數字:'))
        b = int(input('請輸入第二個數字:'))
        break
    except ValueError:
        print('輸入的數值錯誤!請重新輸入!')
print(divide(a,b))
print('完成')
```

```python
def divide(a,b):
    return a/b
    
while True:
    try:
        a = int(input('請輸入第一個數字:'))
        b = int(input('請輸入第二個數字:'))
        break
    except ValueError:
        print('輸入的數值錯誤!請重新輸入!')
try:
    print(divide(a,b))
except ZeroDivisionError:
    print('除數為0!')
print('完成')
```

```python
def divide(a,b):
    return a/b
    
while True:
    try:
        a = int(input('請輸入第一個數字:'))
        b = int(input('請輸入第二個數字:'))
        print(divide(a,b))
        break
    except ValueError:
        print('輸入的數值錯誤!請重新輸入!')
    except ZeroDivisionError:
        print('除以0的錯誤!')

print('完成')
```
