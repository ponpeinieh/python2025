```python
# 輸入文字訊息, 直到'stop' 結束
message = input('請輸入您的文字訊息:')
while message != 'stop':
    print(f'您輸入:{message}')
    message = input('請輸入您的文字訊息:')
print('完成')
```

```python
# 猜數字游戲
import random
count = 0 # 計數器
answer = random.randint(1,100)
guess = int(input('請猜1-100的數字:'))
count += 1
while guess!=answer and count<5:
    if guess > answer:
        print('小一點!')
    elif guess < answer:
        print('大一點!')
    guess = int(input('請猜1-100的數字:'))
    count += 1
if guess == answer :
    print('恭喜您猜對了!')    
else:
    print('您已經猜了5次尚未猜對! 下次再加油!')
```

```python
# Fibonacci 數列
n1 = 0
n2 = 1
while n1<10:
    print(n1, end =' , ')
    mysum = n1 + n2
    n1 = n2
    n2 = mysum
```

```python
# 無窮迴圈 + break (中斷)
message = input('請輸入您的訊息:')
while True:
    print(f'您的訊息: {message}')
    if message == 'stop' :
        break
    message = input('請輸入您的訊息:')
print('完成')
```

```python
# 猜數字游戲
import random
count = 0 # 計數器
answer = random.randint(1,100)
print(answer)
guess = int(input('請猜1-100的數字:'))
count += 1
while guess!=answer:
    if guess > answer:
        print('小一點!')
    elif guess < answer:
        print('大一點!')
    guess = int(input('請猜1-100的數字:'))
    if guess<1 or guess>100: # 不在1-100則continue (不會算做猜了一次)
        continue
    count += 1
    if count == 5:
        break
if guess == answer :
    print('恭喜您猜對了!')    
else:
    print('您已經猜了5次尚未猜對! 下次再加油!')
```

```python
# 列印 1-100但是跳過7的倍數
n = 1
while n<=100:
    if n%7 == 0 :
        n += 1
        continue
    print(n, end = ',')
    n += 1
```

```python
# 宣告變數
deposit = 10000 #本金
rate = 0.03 # 年利率
year = 0 # 年(計數器)
while deposit < 20000:
    deposit += deposit * rate
    year += 1
    print(f'經過{year}年後, 本金為{deposit:.2f}!')
```

```python
# 
heads = int(input("頭的數目:")) #5
legs = int(input("脚的數目:"))
has_answer = False # 是否有解答
for rabbit in range(heads+1):
    chicken = heads - rabbit 
    if rabbit*4 + chicken*2 == legs:
        print(f'兔子:{rabbit}, 雞:{chicken}') 
        has_answer = True
        break
if has_answer == False:
    print('無解答!')
```

```python
# 質數判斷
n = int(input('請輸入一個正整數:'))
is_prime = None # 空值
if n <= 1:
    is_prime = False
elif n<=3:
    is_prime = True
else:
    is_prime = True
    for i in range(2, int(n**0.5)+1):
        print(i)
        if n%i == 0: 
            is_prime = False
            break
print("是質數" if is_prime else "非質數")
```

```python
# 質數判斷
n = int(input('請輸入一個正整數:'))
is_prime = None # 空值
if n <= 1:
    is_prime = False
elif n<=3:
    is_prime = True
else:
    is_prime = True
    i = 2
    while i*i<=n:
        if n%i == 0: 
            is_prime = False
            break
        i+=1
print("是質數" if is_prime else "非質數")
```
