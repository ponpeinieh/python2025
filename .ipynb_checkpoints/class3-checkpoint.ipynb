```python
# 質數判斷
n = int(input('請輸入一個正整數:'))
is_prime = None # 空值
if n <= 1:
    is_prime = False
elif n<=3:
    is_prime = True
else:
    is_prime = True
    for i in range(2, int(n**0.5)+1):
        print(i)
        if n%i == 0: 
            is_prime = False
            break
print("是質數" if is_prime else "非質數")
```

```python
# 質數判斷
n = int(input('請輸入一個正整數:'))
is_prime = None # 空值
if n <= 1:
    is_prime = False
elif n<=3:
    is_prime = True
else:
    is_prime = True
    i = 2
    while i*i<=n:
        if n%i == 0: 
            is_prime = False
            break
        i+=1
print("是質數" if is_prime else "非質數")
```

```python
def bmi(height, weight):
    BMI = weight/(height**2)
    return BMI
# 呼叫
h = 1.8
w = 80
result = bmi(h, w)
print(f'身高{h}公尺, 體重{w}公斤, BMI為{result}')
w = 100
result = bmi(h, w)
print(f'身高{h}公尺, 體重{w}公斤, BMI為{result}')
# 名稱參數
result = bmi(weight=w, height=h) # 順序不重要, 可顛倒
print(f'身高{h}公尺, 體重{w}公斤, BMI為{result}')
```

2.雞兔同籠問題
撰寫一個函數,參數包括總共的頭的數目及脚的數目, 
然後回傳兔子與雞各有幾隻?

3.是否為質數
請寫一個函數，接受一個正整數參數, 回傳是否為質數(True/False)?

```python
'''
1.複利問題
(a)撰寫一個函數, 
參數包含本金, 利率, 年
然後回傳最後的本金
'''
def calc(deposit, rate, year):
    for i in range(year):
        deposit = deposit + deposit*rate
    return deposit
# 呼叫
deposit = 100000
rate = 0.05
year = 10
final_deposit = calc(deposit, rate, year)
print(f'經過{year}年之後, 本金{deposit}, 利率{rate}, 最後本金為{final_deposit:.0f}')
```

```python
#鷄兔同籠
def cage(heads, legs):
    for rabbit in range(heads+1):
        chicken = heads - rabbit 
        if rabbit*4 + chicken*2 == legs:
            # print(f'兔子:{rabbit}, 雞:{chicken}') 
            return (rabbit, chicken)
    else:
        return (-1, -1)
# 呼叫
result = cage(12,36)
print(f'兔子:{result[0]}, 雞:{result[1]}') 
result = cage(10,10)
if result[0]==-1 and result[1]==-1:
    print('無解答')
else:
    print(f'兔子:{result[0]}, 雞:{result[1]}') 
```

額外練習

1. 猜數字游戲擴展
輸入一個游戲難度等級(三個等級: 1. 簡單, 2.一般, 3. 困難)
簡單: 數字範圍 1-100, 最多猜的次數 6 次
一般: 數字範圍 1-1000, 最多猜的次數 9 次
困難: 數字範圍 1-10000, 最多猜的次數 11 次

2. 折扣計算
撰寫一個函數, 參數為原始訂單金額, 根據以下規則打折, 然後回傳折扣後金額
(1) `1~999` : 不打折
(2) `1000~1999` : 95折
(2) `2000~2999` : 9折
(2) `3000~4999` : 85折
(2) `5000`以上 : 8折



 

```python
# 猜數字游戲
import random
# 額外變數
level = int(input('游戲難度等級(1. 簡單, 2.一般, 3. 困難)'))
if level == 1:
    max_value = 100
    max_guess = 6
elif level == 2:
    max_value = 1000
    max_guess = 9
elif level == 3:
    max_value = 10000
    max_guess = 11
count = 0 # 計數器
answer = random.randint(1,max_value)
#print(answer)
guess = int(input(f'請猜1 - {max_value}的數字:'))
count += 1
while guess!=answer:
    if guess > answer:
        print('小一點!')
    elif guess < answer:
        print('大一點!')
    guess = int(input(f'請猜1 - {max_value}的數字:'))
    if guess<1 or guess>max_value: # 不在1-max_value則continue (不會算做猜了一次)
        continue
    count += 1
    if count == max_guess:
        break
if guess == answer :
    print('恭喜您猜對了!')    
else:
    print(f'您已經猜了{max_guess}次尚未猜對! 下次再加油!')
```
