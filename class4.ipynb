```python
def calc_square_and_cube(number):
    square = number**2
    cube = number**3
    return {'square':square, 'cube':cube}

num = 10
result = calc_square_and_cube(num)
print(f'{num}的平方:{result['square']}, 立方:{result['cube']}')
```

```python
# filter() 過濾器
# 隨機產生10個1-100的數字
import random
my_numbers = []
for i in range(10):
    my_numbers.append(random.randint(1,100))
print(my_numbers)
```

```python
#使用filter()篩選奇數的, 偶數丟掉
odd = lambda x : x%2==1
result = list(filter(odd, my_numbers))
print(result)
```

1. 隨機產生10個二維空間的坐標(x坐標, y坐標都介於-10~10之間 ),tuple的形式 
然後儲存到清單中
2. 僅篩選在第一象限的坐標


```python
# 宣告空的清單
import random
coordinates =[]
for _ in range(10):
    pos = (random.randint(-10,10), random.randint(-10,10)) # tuple (元組)
    coordinates.append(pos)
print(coordinates)
```

```python
# 過濾 : 只留下第一象限的位置
f = lambda t: t[0]>0 and t[1]>0
result = list(filter(f, coordinates))
print(result)
```

```python
class User:
    # init -> initialize(初始化) 建構(constructor)函數
    def __init__(self, ID, name, email, age): # 或稱爲方法(method)
        self.ID = ID
        self.name=name
        self.email = email
        self.age = age
    def greet_user(self, message):
        print(f'{message}, {self.name}')
    def validate(self):
        return self.age>=18
# 產生物件
u1 = User(101, 'John Smith', 'j.smith@gmail.com',20)
u2 = User(102, 'Peter Dunn', 'p.dunn@gmail.com',17)
print(f'u1的name: {u1.name}')
print(f'u2的email: {u2.email}')
u1.greet_user('Welcome to python world')
u2.greet_user('Hello')
print(f'u1審核通過? {u1.validate()}')
print(f'u2審核通過? {u2.validate()}')
```

```python
import turtle
# 建立turtle物件
my_turtle = turtle.Turtle()
my_turtle.shape('turtle')
my_turtle.shapesize(2,2)
my_turtle.pencolor('blue')
my_turtle.fillcolor('yellow')
my_turtle.pensize(3)
my_turtle.speed(2)
# 畫圖
my_turtle.begin_fill() # 開始填充
# 正方形
for _ in range(4):
    my_turtle.forward(100)
    my_turtle.right(90)
my_turtle.end_fill() # 結束填充
# 加入第二隻turtle
your_turtle = turtle.Turtle()
your_turtle.shape('circle')
your_turtle.shapesize(2,2)
your_turtle.pencolor('red')
your_turtle.fillcolor('purple')
your_turtle.pensize(3)
your_turtle.speed(1)
your_turtle.penup()
your_turtle.goto(100,200)
your_turtle.pendown()
# 畫三角形    
your_turtle.begin_fill() # 開始填充
for _ in range(3):
    your_turtle.forward(130)
    your_turtle.right(120)
your_turtle.end_fill() # 結束填充
# 完成繪圖
turtle.done() 
try:
    turtle.bye()
except:
    print('good bye')
```
