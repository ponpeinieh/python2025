```python
class User:
    # init -> initialize(初始化) 建構(constructor)函數
    def __init__(self, ID, name, email, age): # 或稱爲方法(method)
        self.ID = ID
        self.name=name
        self.email = email
        self.age = age
    def greet_user(self, message):
        print(f'{message}, {self.name}')
    def validate(self):
        return self.age>=18
# 產生物件
u1 = User(101, 'John Smith', 'j.smith@gmail.com',20)
u2 = User(102, 'Peter Dunn', 'p.dunn@gmail.com',17)
print(f'u1的name: {u1.name}')
print(f'u2的email: {u2.email}')
u1.greet_user('Welcome to python world')
u2.greet_user('Hello')
print(f'u1審核通過? {u1.validate()}')
print(f'u2審核通過? {u2.validate()}')
```

```python
import turtle
# 建立turtle物件
my_turtle = turtle.Turtle()
my_turtle.shape('turtle')
my_turtle.shapesize(2,2)
my_turtle.pencolor('blue')
my_turtle.fillcolor('yellow')
my_turtle.pensize(3)
my_turtle.speed(2)
# 畫圖
my_turtle.begin_fill() # 開始填充
# 正方形
for _ in range(4):
    my_turtle.forward(100)
    my_turtle.right(90)
my_turtle.end_fill() # 結束填充
# 加入第二隻turtle
your_turtle = turtle.Turtle()
your_turtle.shape('circle')
your_turtle.shapesize(2,2)
your_turtle.pencolor('red')
your_turtle.fillcolor('purple')
your_turtle.pensize(3)
your_turtle.speed(1)
your_turtle.penup()
your_turtle.goto(100,200)
your_turtle.pendown()
# 畫三角形    
your_turtle.begin_fill() # 開始填充
for _ in range(3):
    your_turtle.forward(130)
    your_turtle.right(120)
your_turtle.end_fill() # 結束填充
# 完成繪圖
turtle.done() 
try:
    turtle.bye()
except:
    print('good bye')
```

### 練習：英文單字出現次數

- 選擇一首英文歌曲，計算單字及出現次數
  - 忽略大小寫差異，全部轉換為小寫
  - 轉換為list
  - 以dict 儲存
    - 英文單字為key，出現次數為value

```python
lyrics = '''I will survive oh as long as I know how to love I know I will stay alive
I have got all my life to live and I will survive I will survive'''
lower_lyrics = lyrics.lower()
words = lower_lyrics.split()
# print(words)
word_dict = {}
for w in words:
    if w in word_dict:
        word_dict[w] = word_dict[w]+1
    else:
        word_dict[w] = 1
print(word_dict)
```

```python
lyrics = '''I will survive oh as long as I know how to love I know I will stay alive
I have got all my life to live and I will survive I will survive'''
lower_lyrics = lyrics.lower()
words = lower_lyrics.split()
# print(words)
import collections
word_dict = collections.Counter()
for w in words:
    word_dict[w] = word_dict[w]+1
print(word_dict)
```

```python
lyrics = '''I will survive oh as long as I know how to love, I know I will stay alive.
I have got all my life to live and I will survive I will survive.'''
lower_lyrics = lyrics.lower()
words = lower_lyrics.split()
# print(words)
word_dict = {}
for w in words:
    char_list = []
    for c in w:
        if c.isalpha():
            char_list.append(c)
    w = "".join(char_list)    
    if w in word_dict:
        word_dict[w] = word_dict[w]+1
    else:
        word_dict[w] = 1
print(word_dict)
sorted_dict = dict(sorted(word_dict.items(), key=lambda t: len(t[0])))
print(sorted_dict)
```
