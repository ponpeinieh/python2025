```python
# KeyError 鍵值錯誤
def access_dict(dct, key):
    if key in dct:
        return dct[key]
    else:
        return None

my_dict = {'apple':10, 'orange':5, 'grapes':12}
print(access_dict(my_dict, 'apple'))
print(access_dict(my_dict, 'banana'))
```

```python
# 自訂錯誤(例外)型態
class NegativeValueError(Exception):
    pass # 沒有内容
# 丟出(raise 產生)例外的物件

def calc_bmi(height, weight):
    if height<0 or weight<0:
        raise NegativeValueError('身高及體重不能為負數!')
    return weight/height**2

h = float(input('身高:'))
w = float(input('體重:'))
print(calc_bmi(h,w))
```

1. 請改寫猜數字游戲:
(a): 使用try-except處理玩家輸入非數字的情況, 讓玩家重新輸入
(b):  請自訂一個例外型態(類別) - InvalidValueError 
     當玩家輸入的數字不介於1-100之間時, 可以丟出這個錯誤, 並且處理它(讓玩家重新輸入)
 

```python
# 自訂型態
class InvalidValueError(Exception):
    pass
    
# 猜數字游戲
import random
# 額外變數
level = int(input('游戲難度等級(1. 簡單, 2.一般, 3. 困難)'))
if level == 1:
    max_value = 100
    max_guess = 6
elif level == 2:
    max_value = 1000
    max_guess = 9
elif level == 3:
    max_value = 10000
    max_guess = 11
count = 0 # 計數器
answer = random.randint(1,max_value)
#print(answer)
while True:
    try:
        guess = int(input(f'請猜1 - {max_value}的數字:'))
        if guess<1 or guess>max_value: # 不在1-max_value則raise InvalidValueError 
            raise InvalidValueError(f'數值必須在1-{max_value}之間!')
        count += 1
        while guess!=answer:
            if guess > answer:
                print('小一點!')
            elif guess < answer:
                print('大一點!')
            guess = int(input(f'請猜1 - {max_value}的數字:'))
            if guess<1 or guess>max_value: # 不在1-max_value則raise InvalidValueError 
                raise InvalidValueError(f'數值必須在1-{max_value}之間!')
            count += 1
            if count == max_guess:
                break
        if guess == answer :
            print('恭喜您猜對了!')    
        else:
            print(f'您已經猜了{max_guess}次尚未猜對! 下次再加油!')
        break
    except ValueError:
        print('請輸入數字!')
    except InvalidValueError:
        print(f'數值必須在1-{max_value}之間!')
```
