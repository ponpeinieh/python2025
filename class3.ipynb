```python
# while loop 迴圈
count = 0
while count<10:
    print(count, 'hello world!')
    count += 1
```

```python
# 倒數方式
count = 10
while count>0:
    print(count, 'Hello World!')
    count -= 1
```

```python
# 輸入文字訊息, 直到'stop' 結束
message = input('請輸入您的文字訊息:')
while message != 'stop':
    print(f'您輸入:{message}')
    message = input('請輸入您的文字訊息:')
print('完成')
```

```python
# 猜數字游戲
import random
count = 0 # 計數器
answer = random.randint(1,100)
guess = int(input('請猜1-100的數字:'))
count += 1
while guess!=answer and count<5:
    if guess > answer:
        print('小一點!')
    elif guess < answer:
        print('大一點!')
    guess = int(input('請猜1-100的數字:'))
    count += 1
if guess == answer :
    print('恭喜您猜對了!')    
else:
    print('您已經猜了5次尚未猜對! 下次再加油!')
```

```python
# Fibonacci 數列
n1 = 0
n2 = 1
while n1<10:
    print(n1, end =' , ')
    mysum = n1 + n2
    n1 = n2
    n2 = mysum
```

```python
# 無窮迴圈 + break (中斷)
message = input('請輸入您的訊息:')
while True:
    print(f'您的訊息: {message}')
    if message == 'stop' :
        break
    message = input('請輸入您的訊息:')
print('完成')
```

```python
# 列印 1-100但是跳過7的倍數
n = 1
while n<=100:
    if n%7 == 0 :
        n += 1
        continue
    print(n, end = ',')
    n += 1
```
