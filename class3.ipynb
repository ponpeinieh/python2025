```python
# 單行if語法
gender = 'F' # or 'M'
if gender =='F':
    print('女士您好')
else:
    print('先生您好')

message = '女士您好' if gender =='F' else '先生您好'
print(message)
```

```python
# while loop 迴圈
count = 0
while count<10:
    print(count, 'hello world!')
    count += 1
```

```python
# 倒數方式
count = 10
while count>0:
    print(count, 'Hello World!')
    count -= 1
```

```python
# 輸入文字訊息, 直到'stop' 結束
message = input('請輸入您的文字訊息:')
while message != 'stop':
    print(f'您輸入:{message}')
    message = input('請輸入您的文字訊息:')
print('完成')
```

```python
# 猜數字游戲
import random
count = 0 # 計數器
answer = random.randint(1,100)
guess = int(input('請猜1-100的數字:'))
count += 1
while guess!=answer and count<5:
    if guess > answer:
        print('小一點!')
    elif guess < answer:
        print('大一點!')
    guess = int(input('請猜1-100的數字:'))
    count += 1
if guess == answer :
    print('恭喜您猜對了!')    
else:
    print('您已經猜了5次x
```
