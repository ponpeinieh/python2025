```python
# 使用變數(variable)
age = 20
print("我今年",age,"歲!", sep=',') # separator分隔符號
print(f'我今年{age}歲!') # 格式字串語法(formatted string)
print('我今年{age}歲!')
print(f'我明年{age+1}歲!')
```

```python
print(f'我今年{age}歲!', end='\n') # end: 指定列印一行後面的結束符號
print(f'我明年{age+1}歲!')
```
